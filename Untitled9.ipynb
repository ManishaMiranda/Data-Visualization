```python
!pip install folium
import pandas as pd
import folium
covid = folium.Map(location=[27.78778121,-82.34510464], zoom_start=2,tiles='cartodbpositron')
```

```python
Location = r"C:\Users\maria\OneDrive\Documents\Data Visualization\DV COVID-19.xlsx" 
df = pd.read_excel(Location) 
```

```python
df
```

```python
df1 = df.rename(columns={"Admin2": "County", "Province_State": "State", "Country_Region": "Country", "Lat" : "Lattitude", "Long_": "Longitude"})
df1
```

```python
for County,State,Country,Lattitude,Longitude,Confirmed,Deaths,Recovered,Active in zip(df1['County'], df1['State'],df1['Country'],df1['Lattitude'],df1['Longitude'],df1['Confirmed'],df1['Deaths'],df1['Recovered'],df1['Active']):
    folium.CircleMarker([Lattitude, Longitude],
                        radius=Confirmed/1000,
                        color='red',
                         popup =('Place:' + str(County) +' '+ str(State)+' '+ str(Country) + '<br>'
                             'Confirmed Cases:' + str(Confirmed) + '<br>'
                             'Deaths:' + str(Deaths) + '<br>'
                             'Recovered:' + str(Recovered) + '<br>'
                            'Active:' + str(Active) + '<br>'
                             ),
                      
                        fill_color='red',
                        fill_opacity=0.7 ).add_to(covid)
```
