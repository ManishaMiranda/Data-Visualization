```python
import pandas as pd 
Location = r"C:\Users\maria\OneDrive\Documents\Data Visualization\datasets\gradedata.csv" 
df = pd.read_csv(Location) 
df.head()
```

```python
def map_identity(x):
    if str.lower(x) == 'male':
        return 0
    elif str.lower(x) == 'female':
        return 1
    else: 
        return 3

df["Gendervalues"] = df["gender"].map(map_identity)
df.head()
```

```python
import statsmodels.formula.api as sm 
result = sm.ols(       
    formula='grade ~ age + exercise + hours',   
    data=df).fit() 
result.summary()
```

```python
import statsmodels.formula.api as sm 
result = sm.ols(     
    formula='grade ~ exercise + hours',  
    data=df).fit()
result.summary()
```

```python
result = sm.ols(     
    formula='grade ~ Gendervalues + exercise + hours',  
    data=df).fit()
result.summary()
```
